```python
import pandas as pd
import numpy as np

df=pd.read_csv('./hjj_implant_all_metadata_level.csv')
df=df[(df['cohort_num']==1) | (df['cohort_num']== 2)].reset_index(drop=True)
df['path_of_dicom']=''
print(df.__len__)
```

```python
import os
for i in range(len(df)):
    str1=str(df.iloc[i]['anon_dicom_path'])
     
    list1=str1.split('/')
    list2=list1[7:]
    print(list2)
    mylist='F:/new_dataset/images'
    for t in list2:
        mylist=os.path.join(mylist,t)
    df.loc[i,'path_of_dicom']=mylist
    df.loc[i,'anon_dicom_path']=os.path.join('/mnt/new_data',mylist)

df1_p=df.sample(frac=0.02,random_state=42)
df_label=df[df['laterality_label']==1]
new_df=pd.concat([df1_p,df_label],ignore_index=True)
new_df.to_csv('./new_csv.csv',index=False)
```
